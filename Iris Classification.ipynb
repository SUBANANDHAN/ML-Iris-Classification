## Data Visualizations:

## Split the train and test dataset:

### define independent variable/feature variable/input and dependent variable/target variable/output:

```python
X = df.drop(["target","species"],axis = "columns")
y = df.species
```

## Feature Variables:

```python
X.head()
```

## Target Variables:

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=20,random_state = 42,stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Train the models:

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
```

```python
svm_model.fit(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(X_train,y_train)
```

## Test the models:

```python
svm_predict = svm_model.predict(X_test)
```

```python
tree_predict = tree_model.predict(X_test)
```

## Accuracy of the models:

```python
print("Support Vector Machine Accuracy: ",svm_model.score(X_test,y_test) * 100,"%")
```

```python
print("Decision Tree Accuracy: ",tree_model.score(X_test,y_test) * 100,"%")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,svm_predict)
```

```python
flower.target_names
```

```python
sns.heatmap(confusion_matrix(y_test,svm_predict),annot = True,xticklabels=["setosa","versicolor","virginica"],
            yticklabels=["setosa","versicolor","virginica"],cbar=False)
plt.xlabel("Predicted Values")
plt.ylabel("Truth Values")
plt.title("Confusion Matrix Of Support Vector Machine")
plt.show()
```

```python
sns.heatmap(confusion_matrix(y_test,tree_predict),annot = True,xticklabels=["setosa","versicolor","virginica"],
            yticklabels=["setosa","versicolor","virginica"],cbar=False)
plt.xlabel("Predicted Values")
plt.ylabel("Truth Values")
plt.title("Confusion Matrix Of Decision Tree")
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print("Classifiction Report Of Support Vector Machine")
print(classification_report(y_test,svm_predict))
```

```python
print("Classifiction Report Of Decision Tree")
print(classification_report(y_test,tree_predict))
```
